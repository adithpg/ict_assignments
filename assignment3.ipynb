```python
#1 write a python function to find the max of three number
def biggest(x,y,z):
 if x>y and x>z:
    print(x,'is the biggest number')
 elif y>z and y>x:
    print(y,'is the biggest number')
 else:
    print(z,'is the biggest number')




```

```python
x=int(input('enter the first number:'))
y=int(input('enter the second number:'))
z=int(input('enter the third number:'))
biggest(x,y,z)
```

```python
#2 Write a Python function to sum all the numbers in a list
def sum_list_numbers(numbers):
  total = 0
  for x in numbers:
      total+=x
  return total
sample_list =[18,12,30,10,7]
print(f"sample list:{sample_list}")
print(f"expected output:{sum_list_numbers(sample_list)}")
```
