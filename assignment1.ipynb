8.Area of a Triangle (Using Heronâ€™s Formula)

```python
a=float(input('enter the side a:'))
b=float(input('enter the side b:'))
c=float(input('enter the side c:'))
s=(a+b+c)/2
area=(s*(s-a)*(s-b)*(s-c))**0.5
print('area of triangle:',area)
```
